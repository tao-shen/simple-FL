```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import torch
from collections import defaultdict


t = np.arange(0, 100)
fig, axs = plt.subplots(len(a.record['u'][0]), 3, figsize=(15, 15),sharex=True,sharey=True)
fig.set_label('')
for col in range(3):
    alpha=['natural','0.1','0.01']
    a=torch.load('aggr_femnist_alpha_{}.pt'.format(alpha[col]))
    b= defaultdict(list)
    for i in range(len(a.record['u'])):
        for k in a.record['u'][i].keys():
            c=a.record['delta_w'][i][k]*a.record['u'][i][k]*(1-a.p)
            b[k]+=[torch.norm(c.cpu(),p=2)/torch.norm(a.record['delta_w'][i][k].cpu(),p=2)]
    for row,k in zip(range(len(a.record['u'][0])), b):
        points = np.array([t, b[k]]).T.reshape(-1, 1, 2)
        segments = np.concatenate([points[:-1], points[1:]], axis=1)
        norm = plt.Normalize(0, 2.5)
        lc = LineCollection(segments, cmap='viridis', norm=norm)
        # Set the values used for colormapping
        lc.set_array(b[k])
        lc.set_linewidth(2)
        line = axs[row][col].add_collection(lc)
        # axs[row].plot(t, b[k])
        # axs[row].set_xlim(0, 2)
        # axs[0].set_xlabel('time')
        axs[row][col].set_ylabel(k)
        # axs[0].grid(True)
        axs[row][col].set_xlim(0, 100)
        axs[row][col].set_ylim(0, 2.5)

# axs[1].set_ylabel('coherence')
fig.colorbar(line, ax=axs)
# fig.tight_layout()
plt.show()
```
